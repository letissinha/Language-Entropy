```python
#Author: L S Magalhaes (leticiamagalhaes5@gmail.com)
#Date: April 15th, 2019
#Course: 1S2019 - F349 - "Introduo  Teoria de Informao: Aspectos clssicos e qunticos"

#Description : Calculate letter probability and language entropy.
#              The second cell displays a linguistic phylogenetic tree if one wants to analyse
#              proximity.
#
#              Note: when adding languages (to lang_list), make sure the source text uses
#                    latin alphabet.
#                    *Edit lang_list to restrain study for a set of languages
#-----------------------------------------------------------------------------------------------
#                    Available languages:
#                   'portuguese','english', 'afrikaans', 'albanian', 
#                   'catalan', 'czech','danish', 'dutch', 'esperanto','estonian',
#                   'french','finnish','german', 'hungarian', 'indonesian', 
#                   'italian', 'latvian', 'lithuanian', 'norwegian', 'polish',
#                   'romanian', 'serbian', 'slovak', 'slovenian', 'spanish', 'swedish', 
#                   'turkish', 'vietnamese'
#------------------------------------------------------------------------------------------------

#TODO: Print trees with colorcoded leaves to express entropy or "degree of similarity". That'd be cool

#if you want to save figs, set as true:
flag_savefig=False

import string
import collections
from unidecode import unidecode
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter


alphabet='abcdefghijklmnopqrstuvwxyz'

lang_path='src/' #path to source data
lang_list=['portuguese','english', 'afrikaans', 'albanian',  
           'catalan', 'czech','danish', 'dutch', 'esperanto','estonian',
           'french','finnish','german', 'hungarian', 'indonesian', 
           'italian', 'latvian', 'lithuanian', 'norwegian', 'polish',
           'romanian', 'serbian', 'slovak', 'slovenian', 'spanish', 'swedish', 
           'turkish', 'vietnamese']

#linguistic branches - if desirable
west_germanic=['english', 'german', 'dutch', 'afrikaans']
latin=['portuguese', 'italian', 'romanian', 'french', 'spanish', 'catalan']

base_lang=lang_list[0]
colors=['k','#800000', '#DC143C', '#FA8072', '#FF8C00', '#FFD700', '#9ACD32', '#6B8E23', '#006400', '#00FA9A',
        '#48D1CC', '#1E90FF', '#FF7F50', '#2F4F4F', '#000080', '#8B008B', '#BA55D3', '#FFE4C4', '#F4A460', '#7FFFD4', '#FF00FF',
       '#B22222', '#F0E68C', '#AFEEEE', '#A0522D', '#FF69B4', '#FFE4B5', '#C0C0C0']

#--------------------------#
#------- Functions --------#
#--------------------------#

def char_frequency(str1): #Returns dict with letters and its frequencies
    dict = {}   
    for n in alphabet:
        keys=dict.keys()
        dict[n]=0
        
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
    return dict



def log(x): #returns log2 of x
    #had to implement it in case a language had 0 occurences of a given character. 
    logx=np.zeros(len(x))
    for i in range(len(x)):
        if x[i]!=0:
            logx[i]=np.log2(x[i])
        else:
            logx[i]=0
    return logx
    


def init_dict(lang): #Open texts, clean blank spaces and changes char variations 
                     #(such as accents) to its most similar letter
    text=open(lang_path + lang + '.txt', 'r', encoding='utf-8')
    s=unidecode(text.read().lower().replace(' ', '').replace('\n', ''))
    s=s.translate(str.maketrans('', '', string.punctuation))
    return char_frequency(s)



def sort_freq(dictionary): # sorts dictionary according to frequency
    dic=dictionary.copy()
    sorted_alphabet=""
    for i in range(len(alphabet)):    
        temp=max(zip(dic.values(), dic.keys()))[1]  
        sorted_alphabet+=temp
        dic.pop(temp)
    return sorted_alphabet



def prob_entropy(dic, base_alphabet, lang): #return dic with language info, probability list 
                                            #(ordered according to base alphabet) and entropy
    values=[]
    for key in base_alphabet:
        values.append(dic[key])
    prob=np.asarray(values)/sum(values)
    entropy=-sum( prob*log(prob) )
    
    lang_strc={'language':lang, 'prob':prob, 'entropy':entropy}
    return lang_strc


def freq_barplot(n,index, freq, bar_width, opacity, lang, clr):#plots letter frequency barplot
    
    plt.bar(index+n*bar_width, freq, bar_width,
    alpha=opacity,
    color=clr,
    label=lang)

    plt.xticks(index, base)
    plt.tight_layout()
    
    
#--------------------------#
#---------- Main ----------#
#--------------------------#  

#init base and setting variables
base_dic = init_dict(base_lang)
base=sort_freq(base_dic)
t=0
index = np.arange(len(base))*len(lang_list)
fig, ax = plt.subplots(figsize=(12, 5))
bar_width = 0.85
opacity = 0.8

#bar plot - letter probability
entropies=[]
for lang in lang_list:
    dic=init_dict(lang)
    info=prob_entropy(dic, base, lang)    
    freq_barplot(t, index, info['prob'], bar_width, opacity, lang, colors[t])
    entropies.append(info['entropy'])
    t+=1
plt.xlabel('Letter')
plt.ylabel('Probability')
plt.title('Letter Frequency')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
ax.set_xlim(index[0]-3*bar_width, max(index)+(len(lang_list)+2)*bar_width )
plt.show()

if flag_savefig:
    fig.savefig('figs/barplot.png')
    
    
#entropy bar plot 
yticksN=5
fig, ax = plt.subplots(figsize=(6,4.5))
plt.bar(range(len(lang_list)), np.asarray(entropies))
plt.xticks(range(len(lang_list)), lang_list, rotation=90)
ax.set_ylim(min(entropies)-0.1, max(entropies)+0.02)
plt.ylabel('Shannon Entropy')
plt.title('Entropy for 100 most common words in each language')
plt.show()

if flag_savefig:
    fig.savefig('figs/entropy.png')


#probability correlation - 
    #Please change title and base_lang for desired branch accordingly
title='Comparison for all Languages - pt' #'Comparison for West-Germanic languages'
lang_list=lang_list #west_germanic
base_lang=lang_list[0]
base_dic = init_dict(base_lang)
base=sort_freq(base_dic)
base_info=prob_entropy(base_dic, base, base_lang)
x=base_info['prob']

fig, ax = plt.subplots(figsize=(7, 6.5))
plt.plot(x,x)
t=0
for lang in lang_list:
    dic=init_dict(lang)
    info=prob_entropy(dic, base, lang)
    y=info['prob']
    dev=np.std(y-x)
    plt.scatter(x, y, label=lang+' '+str(round(dev, 4)), color=colors[t])
    t+=1
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.xlabel(base_lang+' symbol-character probability')
plt.ylabel('Other language symbol-character probability')
plt.title(title)
plt.show()

if flag_savefig:
    fig.savefig('figs/'+title+'.png')
```
